# Database for finding cool fractals

```python
# Connect to database
conn1 = duckdb.connect("FractalData.duckdb")
```

```python
# Make sure to close connection after
conn1.close()
```

```python
# Create table for first fractal data collection
conn1.execute("CREATE TABLE Data1 (" + ''.join([" Parameter" + str(i) + " float," for i in range(1,55)]) + " Size int, Coolness int);")
```
