```python
import duckdb
```

# Database for finding cool fractals

```python
# Connect to database
conn = duckdb.connect("FractalData.duckdb")
```

```python
# Make sure to close connection after
conn.close()
```

```python
# Create table for first fractal data collection
conn.execute("CREATE TABLE SimpleData (cReal float, cImag float, zbReal float, zbImag float, zsReal float, zsImag float, Coolness float);")
```

```python
conn.execute("COMMENT ON TABLE SimpleData IS 'This table is a simpler version of all_dimensions_data. It only uses six parameters, the coefficients of c, z bar, and z^2. This uses surfaceArea / sqrt(volume) to find coolness'")
```

```python
conn.execute("TRUNCATE SimpleData")
```

```python
conn.execute("SELECT COUNT(*) FROM SimpleData").fetchall()
```
