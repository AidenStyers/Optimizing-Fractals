```python
# Create StartingParameters
conn = duckdb.connect("FractalData.duckdb")
conn.execute("DROP TABLE StartingParameters")
conn.execute("CREATE TABLE StartingParameters ( size INTEGER, surface_area INTEGER, p" + ", p".join([str(i) + " FLOAT" for i in range(52)]) + ");")
conn.execute("COMMENT ON TABLE StartingParameters IS 'This table contains the parameters and resulting size and surface areas for a bunch of randomly selected parameters. " \
             "This will be filtered and used later as starting points for various algorithms.'")
print(conn.execute("DESCRIBE StartingParameters").fetchall())
conn.close()
```
